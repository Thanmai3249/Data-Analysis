Missing Data can occur when no information is provided for one or more items or for a whole unit. Missing Data is a very big problem in a real-life scenarios. so, let's check if there are any null values in the dataset.

We can use isnull() or notnull() functions to check for null values. By using isnull().sum() to check for missing values, we will directly get the counts of missing data in each column.

```python
df.isnull().sum()
```

```python
#Sum of Yearwise population of each continent
continent_df = df.groupby('Continent').sum()
continent_df
```

# Population Growth Rate from the year 1970 to 2022

```python
continent_df[['2022 Population', '2010 Population', '2000 Population', '1990 Population', '1970 Population']].plot(figsize = (20, 20))
```

```python
num_of_countries = df.groupby(['Continent'])['Country'].nunique()
Countries_in__continents = pd.DataFrame(num_of_countries.sort_values(ascending = False))
Countries_in__continents
```

```python
plt.figure(figsize=(6,6))
df.Continent.value_counts().plot(kind = 'pie', autopct='%1.0f%%')
```

We can see that, highest number of countries are there in Africa and least number of countries are there is South America. 

Now, let's see the population in the year 2022 in each continent.

```python
population_by_continent1 = df.groupby(['Continent'])['2022 Population'].sum()
Population_Counts_2022 = pd.DataFrame(population_by_continent1.sort_values(ascending = False))
Population_Counts_2022
```

We can see that, Asia continet is more populated than other countries & Oceania is less populated continent in the year 2022.

```python
population_by_continent = df.groupby(['Continent'])['1970 Population'].sum()
Population_Counts2_1970 = pd.DataFrame(population_by_continent.sort_values(ascending = False))
Population_Counts2_1970
```

```python
Compare = pd.concat([Population_Counts_2022, Population_Counts2_1970], axis = 1)
Compare
```

```python
Compare.plot(figsize= (10, 6))
```

We can see how the population rate increased from 1970 to 2022.
- The continent Oceania is having less growth in population from 1970 to 2022.
- The continents South & Noth America are having equal growth rate in population.
- Europe continent is also having very less population growth rate.
- Africa continent is having more population growth rate when compared to the other 3 continents.
- Asia Continent is having very high population growth rate among the other four continents.

# Area of each continent

```python
continent_Area = df.groupby(['Continent'])['Area (km²)'].sum()
total_area = pd.DataFrame(continent_Area.sort_values(ascending = False))
total_area
```

```python
total_area.plot(figsize = (20, 6))
```

According to area, Asia is the biggest continent in the world & Oceania is the smallest.
